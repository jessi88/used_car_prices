### Explore Categorical Columns

```python
# number of unique values in categorical columns (excluding the 'VIN'-column)

obj_cols = df.drop(columns=['VIN']).select_dtypes(include='object').columns

for col in obj_cols:
    print(col,': ',df[col].nunique())
```

- The `model` column has 29,649 unique values, which we shouldn't take into a model as is. Ways to handle the column are, among others, (a) to drop `model` categories with low counts, (b) drop the entire `model` column, or (c) cluster `model` into broader categories.
- Similarly, the `region`, `state`, and `manufacturer` columns have between 42 and 404 unique values, which are too high and need to be handled.

```python
# Calculate and plot (bar plot) the percentage of unique values in categorical columns.
# Reduce the amount of categories plotted for variables with too many categories.

i=0
n_cols = 2

fig, ax = plt.subplots(math.ceil((len(obj_cols)/n_cols)), n_cols, figsize=(16,25))
fig.suptitle('Percentage of Unique Values in Categorical Columns', fontsize=15, y=1.)

for col in obj_cols:
    df_col = df[col].value_counts(normalize=True).mul(100).rename('Percentage (%)').reset_index()
    
    # reduce the amount of categories plotted for variables with too many categories
    if col in ['region','model']:
        df_col = df_col[df_col['Percentage (%)'] > 0.8]
    if col in ['state','manufacturer']:
        df_col = df_col[df_col['Percentage (%)'] >= 5.]
 
    g = sns.barplot(x='index', y='Percentage (%)', data=df_col, ax=ax[i//n_cols,i%n_cols])
    
    g.set_xticks(g.get_xticks(), fontsize = 12)
    ax[i//n_cols,i%n_cols].set_xticklabels(ax[i//n_cols,i%n_cols].get_xticklabels(), rotation = 90)
    g.set_yticks(g.get_yticks(), fontsize = 12)
    g.set_title(col, fontsize=12)
    g.set_xlabel('')
    g.set_ylabel('Percentage (%)', fontsize = 12)
    
    # add percentages as text to the bars
    for container in g.containers:
        g.bar_label(container, fmt='%.1f%%', fontsize=12)
    bottom, upper = g.get_ylim()
    g.set_ylim(bottom, upper+2)
    plt.subplots_adjust(hspace=5)
    
    i=i+1

# remove the extra, empty last subplot (if applicable)
if len(obj_cols)%2 == 1:
    ax[i//n_cols,i%n_cols].set_axis_off()

fig.tight_layout()
```

- The data has imbalanced categories. We can deal with low-count categories by, among others, (a) dropping them or (b) combining them into larger categories. We decide to combine them into larger categories, which we will do in the Data Preparation section.
- 96.8% of the `title_status` variable has the same value `clean`. Hence, this variable does not provide information to our problem, and we will drop it in the Data Preparation section.
- The categorical variable `condition` is ordinal.
- The categorical variable `cylinders` may be transformed to an integer variable.

Among the four columns that have a too high number of unique values:
- Each category in the `region` column has a count of less than 0.85%. Hence, we will drop this column in the Data Preparation section.
- Similarly, each category in the `model` column has a count of less than 2%. Hence, we will drop this column in the Data Preparation section.
- The `state` and `manufacturer` columns seem to have relevant categories. In order to reduce the high number of unique values, we will cluster these columns into broader categories in the Data Preparation section.

### Handle Imbalanced Categories

In order to reduce the high number of unique values, we will cluster variables into broader categories by combining categories with similar statistics.

#### `cylinders` column

```python
# look at price statistics for each cylinder count
df_clean[df_clean['price']<80000].groupby('cylinders')['price'].describe().sort_values(by=['50%'])
```

We decide to combine:
 - `6` and `10` cylinders
 - `3`, `5`, and `12` cylinders with `other` cylinders

```python
df_clean['cylinders'] = np.where(df_clean['cylinders'].isin(['3 cylinders','5 cylinders','12 cylinders']),'other',
                                df_clean['cylinders'])
df_clean['cylinders'] = np.where(df_clean['cylinders'].isin(['6 cylinders','10 cylinders']),'6_10_cylinders',
                                df_clean['cylinders'])
df_clean['cylinders'].unique()
```

#### Numeric Columns
We compare outlier treatments using (a) the IQR and (b) the z-score.

```python
# distribution plots of numeric columns

num_cols = df_clean.select_dtypes(include='number').columns
fig, ax = plt.subplots(1, len(num_cols), figsize=(16,4))

i=0
for col in num_cols:
    sns.histplot(df_clean[df_clean[col]<1000000][col], ax=ax[i])
    i=i+1
    
fig.tight_layout()
```

```python
# function that checks the percentage of outliers based on the IQR

def detect_outliers_iqr(data, col):
    Q3 = np.quantile(data[col], 0.75)
    Q1 = np.quantile(data[col], 0.25)
    IQR = Q3 - Q1
 
    lower_range = Q1 - 1.5 * IQR
    upper_range = Q3 + 1.5 * IQR
    outlier_list = [x for x in data[col] if ((x < lower_range) | (x > upper_range))]
    
    return 100.*len(outlier_list)/data.shape[0]
```

```python
# function that checks the percentage of outliers based on the z-score

def detect_outliers_zscore(data, col):
    zscore = np.abs(stats.zscore(data[col]))
    outlier_list = np.where(zscore>3,True,False)
    
    return 100.*outlier_list.sum()/data.shape[0]
```

```python
# calculate percentage of outliers using the detect_outliers_iqr() function
outlier_iqr_dic = {}
for col in df_clean.select_dtypes(include='number').columns:
    outlier_val = detect_outliers_iqr(df_clean.dropna(subset=col), col)
    if outlier_val > 0:
        outlier_iqr_dic[col] = outlier_val
```

```python
# calculate percentage of outliers using the detect_outliers_zscore() function
outlier_zscore_dic = {}
for col in df_clean.select_dtypes(include='number').columns:
    outlier_val = detect_outliers_zscore(df_clean.dropna(subset=col), col)
    if outlier_val > 0:
        outlier_zscore_dic[col] = outlier_val
```

```python
# sort the columns by their percentage of outliers (IQR based)
outlier_iqr_dic_sorted = {}
sorted_keys = sorted(outlier_iqr_dic, key=outlier_iqr_dic.get, reverse=True)

for w in sorted_keys:
    outlier_iqr_dic_sorted[w] = outlier_iqr_dic[w]
    
# plot percentage of outliers
plt.figure(figsize=(16, 1))
sns.barplot(x=list(outlier_iqr_dic_sorted.values()), y=list(outlier_iqr_dic_sorted.keys()));
```

```python
# sort the columns by their percentage of outliers (z-score based)
outlier_zscore_dic_sorted = {}
sorted_keys = sorted(outlier_zscore_dic, key=outlier_zscore_dic.get, reverse=True)

for w in sorted_keys:
    outlier_zscore_dic_sorted[w] = outlier_zscore_dic[w]
    
# plot percentage of outliers
plt.figure(figsize=(16, 1))
sns.barplot(x=list(outlier_zscore_dic_sorted.values()), y=list(outlier_zscore_dic_sorted.keys()));
```

### Define Pipeline Preprocessor

We will use the same preprocessor steps for all models.

```python
# Column transformer for numeric variables:
# As mentioned earlier, we use the IterativeImputer to impute missing values.
# We also standardize numerical features.
numeric_transformer = Pipeline(
    steps=[
        ('imputer', IterativeImputer(n_nearest_features=3, sample_posterior=True)),
        ('scaler', StandardScaler())
    ]
)

# Column transformer for ordinal variables:
# We encode ordinal features as an integer array.
# As mentioned earlier, we use the IterativeImputer to impute missing values.
ordinal_transformer = Pipeline(
    steps=[
        ('encoder', OrdinalEncoder(categories = [[np.nan, 'fair_salvage', 'good', 'excellent', 'like new']])),
        ('imputer', IterativeImputer(n_nearest_features=3, sample_posterior=True))
    ]
)    

# Column transformer for nominal variables:
# We use target encoding for nominal features.
# As mentioned earlier, we use the IterativeImputer to impute missing values.
nominal_transformer = Pipeline(
    steps=[
        ('encoder', TargetEncoder()),
        ('imputer', IterativeImputer(n_nearest_features=3, sample_posterior=True))
    ]
)    


# combine the column transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('numerical', numeric_transformer, num_cols),
        ('ordinal', ordinal_transformer, ordinal_cols),
        ('nominal', nominal_transformer, nominal_cols)
    ]
)
```

Next, we will train three different models, and evaluate them based on their R2 score:
- Ridge Regression
- Lasso Regression
- Random Forest Regression

We tune hyperparameters for each model using GridSearchCV (for Ridge and Lasso Regression) and BayesSearchCV (for Random Forest Regression).

For all three models, we use the TransformedTargetRegressor to apply a log-transformation to the target.

We noticed that using `n_estimators=100` has almost the same quality but is quite faster. Hence, we decide to choose the Random Forest Regressor with `max_depth=30` as our final model.

Below, we will train our chosen final model using the data from the training and validation sets, and then we evaluate it using the test data based on different metrics.

```python
# build pipeline
pipe = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('rf', TransformedTargetRegressor(regressor=RandomForestRegressor(max_depth=30),
                                          func=np.log, inverse_func=np.exp))
    ]
)

# fit on train
pipe.fit(X_train, y_train)

# predict on train and test
y_pred_train = pipe.predict(X_train)
y_pred_test = pipe.predict(X_test)
```
